```python
from google.colab import files
from pathlib import Path

uploaded = files.upload()  # Upload your video file
video_path = list(uploaded.keys())[0]
video_path
```

```python
import ffmpeg
import os

def extract_audio(video_path):
    audio_path = Path(video_path).with_suffix('.wav')
    ffmpeg.input(video_path).output(
        str(audio_path),
        **{'q:a': 0, 'map': 'a'}
    ).run(overwrite_output=True)
    return str(audio_path)

audio_output = extract_audio(video_path)
print(f"ğŸ§ Audio saved as: {audio_output}")
```

```python
def convert_to_reel_format(input_video):
    output_path = Path(input_video).stem + "_reel.mp4"
    ffmpeg.input(input_video).output(
        output_path,
        vf="scale=1080:-2,pad=1080:1920:(ow-iw)/2:(oh-ih)/2"
    ).run(overwrite_output=True)
    return output_path

reel_video_path = convert_to_reel_format(video_path)
print(f"ğŸ“± Reel formatted video saved as: {reel_video_path}")
```

```python
# Download main outputs
files.download(audio_output)
files.download(reel_video_path)

# If chunked, zip and download
# !zip -r chunks.zip chunks
# files.download("chunks.zip")
```

```python
import whisper

# Load a lightweight model (options: tiny, base, small, medium, large)
model = whisper.load_model("base")

# Transcribe the audio
result = model.transcribe(audio_output)

# Display full transcript
print("ğŸ“„ Full Transcript:\n")
print(result["text"])
```
